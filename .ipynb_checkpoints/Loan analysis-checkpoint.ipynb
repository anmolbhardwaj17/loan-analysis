Removing columns which are think are not going to help with the analysis

```python
df = df.drop([
  "delinq_2yrs",
  "earliest_cr_line",
  "inq_last_6mths",
  "revol_bal",
  "revol_util",
  "out_prncp",
  "out_prncp_inv",
  "total_rec_prncp",
  "total_rec_int",
  "total_rec_late_fee",
  "recoveries",
  "last_pymnt_d",
  "last_credit_pull_d",
  "application_type"], axis=1)
```

Loan status are of string type but shoud be in cateory as there are only 3 outputs, Fully Paid, Charged off, and Current

```python
df['loan_status'] = df['loan_status'].astype('category')
df['loan_status'].value_counts()
```

Intially I created 3 categories 0 for fully paid, 1 for Charged Off and 2 for current. But later I realised that I have to create a report for Fully paid and current plus 3 categories were giving issues in making a graph

```python
# funcc = lambda x: 0 if x=='Fully Paid' else (1 if x=='Charged Off' else 2)
# df['loan_status'] = df['loan_status'].apply(funcc)
df['loan_status'] = df['loan_status'].apply(lambda x: 0 if x=='Charged Off' else 1)
df['loan_status'] = df['loan_status'].apply(lambda x: pd.to_numeric(x))
```
